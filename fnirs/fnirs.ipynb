## Medium

```python
from problems.two_balls import two_balls_2d_sensors, two_balls_2d_medium

contrast = 1.1
voxels_per_dim = 160
r_ball_mm = 2
depth_mm = 10
ball_separation_mm = 10

medium_bg, medium_true = two_balls_2d_medium(
    contrast, voxels_per_dim, r_ball_mm, depth_mm, ball_separation_mm
)

noptodes = 20
sensors = two_balls_2d_sensors(noptodes, medium_bg)

medium_true.plot_mua(sensors)
```

## Forward

```python
from solver import Solver, jacobian

solver = Solver(medium_bg, sensors, tstart=0, tend=5e-9, tstep=5e-9)
```

```python
nx, ny, nz = medium_bg.nx, medium_bg.ny, medium_bg.nz
ndet, nsrc = sensors.ndet, sensors.nsrc
nt = solver.nt
```

```python
# initialize arrays
J = np.zeros((nz, ny, nx, nt, ndet, nsrc))
data_bg = np.zeros((nt, ndet, nsrc))
data_true = np.zeros((nt, ndet, nsrc))
```

```python
t = np.linspace(solver.tstart, solver.tend, nt)

nphoton = 1e5

# save the data
params = {
    "nphoton": nphoton,
    "noptodes": noptodes,
    "nx": nx,
    "ny": ny,
    "nz": nz,
    "nt": nt,
    "tstart": solver.tstart,
    "tend": solver.tend,
    "tstep": solver.tstep,
    "r_ball_mm": r_ball_mm,
    "depth_mm": depth_mm,
    "ball_separation_mm": ball_separation_mm,
    "contrast": contrast,
}
```

```python
from tqdm.notebook import tqdm

# run forward for all sources (for both the ground truth and background)
for src_idx in tqdm(range(nsrc), desc="Processing sources"):

    solver.medium = medium_bg
    res_bg, cfg_bg = solver.forward(src_idx, random_seed=1, nphoton=nphoton)
    data_bg[..., src_idx] = solver.get_td_data(res_bg, medium_bg.optical_properties)
    J[..., src_idx] = jacobian(res_bg, cfg_bg)
```

## SVD

```python
# get the condition number of J
J_c = J.reshape((nz*ny*nx*nt, ndet*nsrc))
cond = np.linalg.cond(J_c)
print(f"Condition number of J: {cond}")
```

```python
J_c.shape
```

```python
import cupy as cp

# Move data to GPU and compute SVD
J_gpu = cp.asarray(J_c)  # Move to GPU
u, s, vh = cp.linalg.svd(J_gpu, full_matrices=False)
s = cp.asnumpy(s)  # Move results back to CPU for plotting
```

```python
plt.figure()
plt.semilogy(s, '.')
plt.xlabel('Index')
plt.ylabel('Singular value')
plt.title('Singular value spectrum of J')
plt.grid(True)
```
